```python
textgen_word_bigger_4 = textgenrnn(name="brandenbursch_medium_model_4")

textgen_word_bigger_4.reset()
textgen_word_bigger_4.train_on_texts(all_names,
                             new_model=True,
                             rnn_bidirectional=True,
                             word_level=False,
                             max_length=4)

print(textgen_word_bigger_4.model.summary())
```
